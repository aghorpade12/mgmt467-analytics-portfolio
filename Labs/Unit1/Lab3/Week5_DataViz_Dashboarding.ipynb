### Q2 ‚Äî Chart Types & When to Use Them
**Prompt:** *"Describe 5 common chart types (e.g., bar, line, scatter, histogram, map) and the type of data analysis question each is best suited to answer."*  

For each chart, list: **Question archetype**, **Data shape**, **Pitfalls**, **Alternatives**.

üìù **Your response (template):**
1. **Bar:**  
   - <u>Question archetype:</u> Comparing quantities across different categories.
   - <u>Data shape:</u> Primarily uses categorical data for one axis (e.g., product names, months, categories) and numerical data for the other axis (e.g., sales figures, visit counts, frequencies).
   - <u>Pitfalls:</u> Too many categories, misleading y-axis scaling, overlapping bars, used for continuous data
   - <u>Alternatives:</u> Grouped Bar Chart, Stacked Bar Chart, Dot Plot, Treemap
2. **Line:**
   - <u>Question archetype:</u> Showing trends, patterns, or changes over a continuous interval, typically time.
   - <u>Data shape:</u> Requires at least one continuous or ordered variable (e.g., time, dates, sequential measurements) and one numerical variable.
   - <u>Pitfalls:</u> Too many lines, Choppy or irregular intervals, Using for categorical data, Misleading y-axis scaling
   - <u>Alternatives:</u> Area Chart, Sparklines, Slopegraph
3. **Scatter:**
   - <u>Question archetype:</u> Showing the relationship or correlation between two numerical variables. Identifying clusters, outliers, and the strength and direction of a relationship.
   - <u>Data shape:</u> Uses two numerical variables, one plotted on the x-axis and the other on the y-axis. Each point represents a single observation.
   - <u>Pitfalls:</u> Overplotting, Correlation vs. Causation, Ignoring outliers, Used for non-numerical data
   - <u>Alternatives:</u> Bubble Chart, Heatmap, Pair Plot (with histograms/density plots)
4. **Histogram:**
   - <u>Question archetype:</u> Showing the distribution of a single numerical variable. Understanding the frequency of data points within specific bins or intervals.
   - <u>Data shape:</u> Uses a single numerical variable, which is divided into a series of bins (intervals). The height of each bar represents the count (frequency) of data points that fall within that bin.
   - <u>Pitfalls:</u> Bad bin size, Misleading y-axis, Comparing multiple distributions
   - <u>Alternatives:</u> Box Plot, Density Plot (KDE), Frequency Polygon
5. **Map:**
   - <u>Question archetype:</u> Visualizing data geographically, showing spatial patterns, distributions, or comparisons across regions.
   - <u>Data shape:</u> Typically involves geographical data (regions, countries, cities, coordinates) and associated numerical or categorical data that is then overlaid onto the map.
   - <u>Pitfalls:</u> Geographic scale bias, Poor color choice, Overcrowding, Projection distortion
   - <u>Alternatives:</u> Table with geographic sorting, Heatmap on a Grid, Geographic Scatter Plot with Tooltips, Cartograms

### Q3 ‚Äî Airline Executive Dashboard KPIs
**Prompt:** *"Your goal is to create a dashboard for an airline executive. Brainstorm 5 Key Performance Indicators (KPIs) that should be on this dashboard. For each KPI, suggest the best chart type to visualize it."*

Suggested KPI ideas to consider (customize them): On-Time Rate, Average Delay Minutes, Cancellation Rate, Load Factor, Yield (Revenue per Passenger Mile), Completion Factor, Net Promoter Score (if available), Revenue, Cost per Available Seat Mile (CASM).

üìù **Your response (template):**
1. KPI: On-Time Performance (OTP)
   - <u>Why it matters:</u> A crucial indicator of operational efficiency and customer satisfaction. High OTP leads to better customer experience, fewer complaints, and operational cost savings (less disruption). Executives need to monitor trends and identify potential issues impacting punctuality.
   - <u>Best chart:</u>
       - Primary View: Gauge Chart or Large Number Display: To show the current on-time percentage (e.g., last month, last week) as a quick, high-level status.
      - Trend View: Line Chart: To track OTP over time (daily, weekly, monthly). This helps identify seasonal patterns or recent degradations.
      - Breakdown View: Stacked Bar Chart: To show the breakdown of "On-Time," "Delayed," and "Canceled" flights for a given period, allowing comparison of these segments.
   - <u>Drill-down fields:</u>
        - Route (e.g., JFK-LAX)
        - Aircraft Type
        - Day of Week
        - Time of Day
        - Reason for Delay (if available)
        - Origin Airport
        - Destination Airport
2. KPI: Load Factor
   - <u>Why it matters:</u> Measures how efficiently an airline is filling its aircraft. A higher load factor generally indicates better revenue generation per flight, as more seats are occupied and paid for. Executives need to understand how well flights are selling across different routes and times.
   - <u>Best chart:</u>
       - Primary View: Large Number Display with Trend Indicator: Showing the current overall load factor and a small arrow indicating if it's up or down from the previous period.
      - Route Performance: Bar Chart: To compare load factors across different key routes or route segments.
      - Time-based Trends: Line Chart: To track load factor over time, potentially by day of week or season.
   - <u>Drill-down fields:</u>
        - Route
        - Flight Number
        - Date
        - Aircraft Type
        - Booking Channel
        - Seasonality (e.g., Summer vs. Winter)
3. KPI: Yield (Revenue per Passenger Mile - RPM)
   - <u>Why it matters:</u> This KPI reflects the average revenue generated by the airline for each mile flown by a passenger. It's a key profitability metric. Executives want to see if pricing strategies are effective and how yield is performing across different markets.
   - <u>Best chart:</u>
       - Primary View: Large Number Display: Showing the current yield (e.g., cents per mile) with a trend indicator.
      - Route Performance: Bar Chart: To compare yield across different routes or route types (e.g., domestic vs. international).
      - Time-based Trends: Line Chart: To track yield over time, helping to understand the impact of pricing adjustments or market conditions.
   - <u>Drill-down fields:</u>
        - Route
        - Passenger Class (e.g., Economy, Business, First)
        - Date
        - Booking Period (e.g., how far in advance the ticket was booked)
        - Ancillary Revenue (if separate)
4. KPI: Cancellation Rate
   - <u>Why it matters:</u> High cancellation rates negatively impact customer trust, lead to significant operational costs (rebooking, compensation), and damage reputation. Executives need to monitor this closely to ensure operational stability.
   - <u>Best chart:</u>
       - Primary View: Gauge Chart or Large Number Display: Showing the current cancellation rate (percentage of flights canceled).
      - Trend View: Line Chart: To track cancellation rate over time (daily, weekly, monthly).
      - Breakdown View: Bar Chart or Pie Chart: To show cancellations by reason (e.g., weather, technical, crew, air traffic control).
   - <u>Drill-down fields:</u>
        - Route
        - Origin Airport
        - Destination Airport
        - Aircraft Type
        - Date
        - Reason for Cancellation
5. KPI: Cost Per Available Seat Mile (CASM)
   - <u>Why it matters:</u> This is a critical efficiency metric that measures the airline's cost of providing air transportation. Lower CASM generally means higher profitability. Executives must monitor this to ensure cost controls are effective and to benchmark against competitors.
   - <u>Best chart:</u>
       - Primary View: Large Number Display: Showing the current CASM, ideally with a comparison to a target or previous period.
      - Trend View: Line Chart: To track CASM over time, identifying increases or decreases.
      - Breakdown View: Stacked Bar Chart: To show the major components of CASM (e.g., fuel costs, labor costs, maintenance costs, aircraft ownership costs) as a proportion of the total.
   - <u>Drill-down fields:</u>
        - Route (if costs vary significantly by route due to factors like airport fees)
        - Aircraft Type (different aircraft have different operating costs)
        - Month/Quarter (to see cost trends)
        - Cost Category (e.g., Fuel, Labor, Maintenance, Landing Fees)

### Create a Synthetic Airline Dataset (if you don't have one)
Use this dataset to prototype visualizations and dashboards. It contains realistic columns for the KPIs you defined. You can always swap in your BigQuery data later.


````python
# Generate a synthetic airline dataset
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

np.random.seed(467)

dates = pd.date_range(start='2024-01-01', end='2024-06-30', freq='D')
carriers = ['AA', 'DL', 'UA', 'WN', 'AS']
origins = ['ATL','DFW','DEN','ORD','LAX','SEA','SFO','PHX','CLT','MCO']
dests =   ['JFK','LGA','BOS','SFO','LAX','SEA','MIA','IAH','IAD','PHL']

rows = []
for d in dates:
    for carrier in carriers:
        # number of flights per day per carrier
        n = np.random.randint(80, 140)
        for _ in range(n):
            o = np.random.choice(origins)
            de = np.random.choice(dests)
            if o == de:
                continue
            distance = np.random.randint(300, 2800)
            seats = np.random.choice([90,120,150,180,210])
            load_factor = np.clip(np.random.normal(0.82, 0.07), 0.55, 0.98)
            pax = int(seats * load_factor)
            dep_delay = int(np.random.normal(8, 15))
            arr_delay = dep_delay + int(np.random.normal(0, 12))
            dep_delay = max(dep_delay, -5)
            arr_delay = max(arr_delay, -10)
            cancelled = np.random.rand() < 0.025
            diverted = np.random.rand() < 0.005
            revenue = pax * np.random.uniform(120, 420)
            casm = np.random.uniform(0.08, 0.16)  # cost per available seat mile
            cost = casm * seats * distance
            fuel_burn = distance * np.random.uniform(3.0, 4.5)  # gallons per seat-mile proxy
            nps = np.random.normal(35, 15)  # if you want a customer metric proxy
            rows.append({
                'date': d,
                'carrier': carrier,
                'origin': o,
                'dest': de,
                'distance_miles': distance,
                'seats': seats,
                'passengers': pax,
                'dep_delay_min': dep_delay,
                'arr_delay_min': arr_delay,
                'cancelled': int(cancelled),
                'diverted': int(diverted),
                'revenue_usd': round(revenue, 2),
                'cost_usd': round(cost, 2),
                'fuel_burn_gal': round(fuel_burn, 1),
                'nps': round(nps, 1)
            })

df_flights = pd.DataFrame(rows)

# Derived KPIs
df_flights['on_time'] = (df_flights['arr_delay_min'] <= 15).astype(int)
df_flights['load_factor'] = (df_flights['passengers'] / df_flights['seats']).round(3)
df_flights['yield_usd_per_pax_mile'] = (df_flights['revenue_usd'] / (df_flights['passengers'] * df_flights['distance_miles']).clip(lower=1)).round(4)
df_flights['asm'] = df_flights['seats'] * df_flights['distance_miles']  # available seat miles
df_flights['casm'] = (df_flights['cost_usd'] / df_flights['asm']).round(4)
df_flights['month'] = df_flights['date'].dt.to_period('M').astype(str)

display(df_flights.head())
print(df_flights.shape)
````

````python
# --- Code for Prompt A ---
import matplotlib.pyplot as plt
import seaborn as sns


# Set a clean plot style
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))

# Compute on-time rate by carrier
on_time_rate = df_flights.groupby('carrier')['on_time'].mean().sort_values(ascending=False)

# Create the bar plot
ax = sns.barplot(x=on_time_rate.index, y=on_time_rate.values, palette='viridis', hue=on_time_rate.index, legend=False)

# Add percentage labels on each bar
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1%}',
                  (p.get_x() + p.get_width() / 2., p.get_height()),
                  ha='center', va='center',
                  xytext=(0, 9),
                  textcoords='offset points')

# Set labels and title
plt.title('On-Time Performance Rate by Carrier', fontsize=16)
plt.xlabel('Carrier', fontsize=12)
plt.ylabel('On-Time Rate', fontsize=12)
plt.ylim(0.62, 0.69) # Adjusted to better show difference between carriers

# Ensure layout is clean
plt.tight_layout()

# Display the plot
plt.show()
````

````python
# --- Code for Prompt B ---
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")
plt.figure(figsize=(14, 7))

# Aggregate to daily average arrival delay
daily_delay = df_flights.groupby('date')['arr_delay_min'].mean()

# Calculate the 7-day rolling average
rolling_avg_delay = daily_delay.rolling(window=7).mean()

# Plot the daily average and the rolling average
plt.plot(daily_delay.index, daily_delay.values, label='Daily Average Delay', alpha=0.5)
plt.plot(rolling_avg_delay.index, rolling_avg_delay.values, label='7-Day Rolling Average', color='red', linewidth=2)

# Add labels, title, and grid
plt.title('Average Arrival Delay Over Time', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Average Arrival Delay (minutes)', fontsize=12)
plt.grid(True)
plt.legend()

# Ensure layout is clean
plt.tight_layout()

# Display the plot
plt.show()
````

````python
# --- Code for Prompt C ---
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")
plt.figure(figsize=(12, 7))

# Define the data to be plotted, excluding extreme outliers for better visualization
delay_data = df_flights['arr_delay_min'].clip(-30, 120)

# Plot the histogram with appropriate bins
sns.histplot(delay_data, bins=50, kde=False)

# Add a vertical reference line at 0
plt.axvline(x=0, color='red', linestyle='--', linewidth=2, label='On-Time (0 min delay)')

# Calculate and annotate the percentage of flights within the [-5, 15] minute window
on_time_ish_flights = df_flights['arr_delay_min'].between(-5, 15).sum()
total_flights = len(df_flights)
percentage = (on_time_ish_flights / total_flights) * 100

plt.annotate(f'{percentage:.1f}% of flights arrived between\n-5 and +15 minutes of schedule',
             xy=(15, 0), xytext=(30, 10000),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, ha='left')

# Set labels and title
plt.title('Distribution of Arrival Delays', fontsize=16)
plt.xlabel('Arrival Delay (minutes)', fontsize=12)
plt.ylabel('Number of Flights', fontsize=12)
plt.legend()

# Ensure layout is clean
plt.tight_layout()

# Display the plot
plt.show()
````

````python
# --- Code for Prompt D ---
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")
plt.figure(figsize=(12, 8))

# Create a 'route' column for easier grouping
df_flights['route'] = df_flights['origin'] + '-' + df_flights['dest']

# Compute mean load factor by route, get the top 15, and sort for plotting
top_15_routes = df_flights.groupby('route')['load_factor'].mean().nlargest(15).sort_values(ascending=False)

# Create the horizontal bar plot
ax = sns.barplot(x=top_15_routes.values, y=top_15_routes.index, palette='plasma', hue=top_15_routes.index, legend=False)

# Add labels and title
plt.title('Top 15 Routes by Average Load Factor', fontsize=16)
plt.xlabel('Average Load Factor', fontsize=12)
plt.ylabel('Route', fontsize=12)

# Adjust x-axis to better show differences between top routes
plt.xlim(left=top_15_routes.values.min() * 0.999, right=top_15_routes.values.max() * 1.001)

# Ensure layout is clean so labels are not cut off
plt.tight_layout()

# Display the plot
plt.show()

````

````python
# --- Code for Prompt E ---
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")
plt.figure(figsize=(10, 8))

# Create the scatter plot, using a sample to avoid overplotting
sample_df = df_flights.sample(n=5000, random_state=42)

scatter_plot = sns.scatterplot(
    data=sample_df,
    x='cost_usd',
    y='revenue_usd',
    hue='casm',
    palette='coolwarm', # Using a reversed colormap so high CASM is red
    alpha=0.7
)

# Add a y=x reference line (where cost equals revenue)
max_val = max(sample_df['cost_usd'].max(), sample_df['revenue_usd'].max())
plt.plot([0, max_val], [0, max_val], color='gray', linestyle='--', label='Breakeven (Cost = Revenue)')

# Add labels, title, and legend
plt.title('Cost vs. Revenue per Flight (Colored by CASM)', fontsize=16)
plt.xlabel('Cost (USD)', fontsize=12)
plt.ylabel('Revenue (USD)', fontsize=12)
plt.legend()

# Ensure layout is clean
plt.tight_layout()

# Display the plot
plt.show()
````
